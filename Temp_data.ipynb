```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('HadCRUT.5.1.0.0.analysis.summary_series.global.annual.csv')
df = df.rename(columns={'Time': 'year', 'Anomaly (deg C)': 'Temp anomaly'})
df = df[df['year'] >= 1865]


Ys_20 = []
Average_Temp = []

```

```python
def Average_temp_20(df):
    return df['Temp anomaly'].mean()


Average_Temp = []
Ys_20 = []

for i in range(8):
    df_20 = df[(df['year'] > 1865 +20*i) & (df['year'] <= 1865 + 20*(i+1))]
    Average_Temp.append(Average_temp_20(df_20))
    Ys_20.append(f"{1865 + 20*i}")

df_AvgTemp20 = pd.DataFrame({
    "decade": Ys_20,        
    "Average_Temperature": Average_Temp
})
df_AvgTemp20

plt.plot(df_AvgTemp20['decade'], df_AvgTemp20['Average_Temperature'], marker='o')
plt.xlabel('Decade Starting Year')
plt.ylabel('Average Global Temperature Anomaly (Â°C)')
plt.title('Average Global Temperature Anomaly by Decade (1865-2025)')
plt.grid()
plt.xticks(rotation=45)
plt.show()
```
