```python
import pandas as pd
import matplotlib.pyplot as plt

dfTemp = pd.read_csv('HadCRUT.5.1.0.0.analysis.summary_series.global.annual.csv')
dfTemp = dfTemp.rename(columns={'Time': 'year', 'Anomaly (deg C)': 'Temp anomaly'})
dfTemp = dfTemp[dfTemp['year'] >= 1865]




```

```python
def Average_temp(df):
    return dfTemp['Temp anomaly'].mean()
plt.style.use('ggplot')

Average_Temp_30 = []
Ys_30 = []

for i in range(5):
    dfTemp_30 = dfTemp[(dfTemp['year'] > 1875 +30*i) & (dfTemp['year'] <= 1875 + 30*(i+1))]
    Average_Temp_30.append(Average_temp(dfTemp_30))
    Ys_30.append(f"{1875 + 30*i}")

df_AvgTemp30 = pd.DataFrame({
    "decade": Ys_30,        
    "Average_Temperature": Average_Temp_30
})
df_AvgTemp30

plt.plot(df_AvgTemp30['decade'], df_AvgTemp30['Average_Temperature'], marker='o')
plt.xlabel('30 Year Time Frame Beginning')
plt.ylabel('Average Global Temperature Anomaly (Â°C)')
plt.title('Average Global Temperature Anomaly by 30 year period (1875-2025)')
plt.grid()
plt.xticks(rotation=45)
plt.show()
```
